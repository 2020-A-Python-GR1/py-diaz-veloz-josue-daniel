```python
import matplotlib.pyplot as plt
import os
import numpy as np
from scipy import misc

```

```python
def validar_posicion(msg,maximo):
    respuesta = input(msg)
    if(int(respuesta)<int(maximo)):
        return respuesta
    else:
        print("Ingrese una posicion valida")
        return validar_posicion(msg,maximo)
```

```python
mapache = misc.face()
tamanio_tablero_n = input("Ingrese un numero par de division del tablero")
mapache_parts = np.vsplit(mapache,int(tamanio_tablero_n))

piezas_index = np.random.permutation(int(tamanio_tablero_n))
mapache_random = np.concatenate(list(map(lambda x: mapache_parts[x],piezas_index)))

print(" ROMPECABEZAS MAPACHE A RESOLVER")
plt.imshow(mapache_random)
plt.show()

def menu():
    os.system('clear')
    print ("Selecciona una opción")
    print ("\t1 - Jugar")
    print ("\t2 - Salir")

while True:
    menu()
    opcionMenu = input("Ingresa una opcion: ")
    
    if opcionMenu=="1":
        print("STAR GAME")
        mover_posicion_a = validar_posicion("Posicion de la imagen a mover: ",tamanio_tablero_n)
        mover_posicion_b = validar_posicion("Posicion de la imagen nueva: ",tamanio_tablero_n)
        pieza_num = piezas_index[int(mover_posicion_a)]
        piezas_index[int(mover_posicion_a)] = piezas_index[int(mover_posicion_b)]
        piezas_index[int(mover_posicion_b)] = pieza_num
        
        mapache_desordenado = np.concatenate(list(map(lambda x: mapache_parts[x],piezas_index)))
        plt.imshow(mapache_desordenado)
        plt.show()
        
    
    elif opcionMenu=="2":
        print("Salio del rompecabezas")
        break
```
