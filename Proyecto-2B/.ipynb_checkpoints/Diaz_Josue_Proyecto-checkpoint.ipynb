```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Abrir el archivo csv
path = "C:/Users/josh_/Documents/GitHub/py-diaz-veloz-josue-daniel/Proyecto-2B/metadatosHomicidiosAsesinatos.csv"

datos = pd.read_csv(path, sep=';')
```

**1.Analizar el numero de homicidios y asesinatos segun el genero**

```python
p = sns.barplot(x='Sexo', y='Numero_asesinatos', data=datos)
```

**2. Determinar el porcentaje de homicidios y asesinatos segun el genero.**

```python
serie_homicidios_sexo=datos.Sexo.value_counts()
patches,texts=plt.pie(serie_homicidios_sexo,shadow=True,startangle=90)
plt.pie(serie_homicidios_sexo,shadow=True,labels=serie_homicidios_sexo.index,autopct='%1.1f%%')
plt.axis("equal")
plt.title("Homicidios según el sexo")
plt.tight_layout()
plt.show()
```

**3.Analizar el numero de homicidios y asesinatos segun el estado civil**

```python
p = sns.barplot(x='Estado Civil', y='Numero_asesinatos', data=datos)
```

**4.Analizar el numero de homicidios y asesinatos segun el estado civil y genero.**

```python
p = sns.barplot(x='Sexo', y='Numero_asesinatos', hue='Estado Civil', data=datos)
```

**5.Determinar las provincias y sus indice de homicidios y asesinatos**

```python
num_asesinatos = datos['Numero_asesinatos'].groupby(datos.Provincia).mean().sort_values(ascending=False)
f,ax = plt.subplots(1,1,figsize=(15,10))
ax = sns.barplot(num_asesinatos,num_asesinatos.index,palette='coolwarm')
```

**6.Determinar las provincias con mayor indice de homicidios y asesinatos.**

```python
num_asesinatos = datos['Numero_asesinatos'].groupby(datos.Provincia).mean().sort_values(ascending=False)
f,ax = plt.subplots(1,1,figsize=(15,10))
ax = sns.barplot(num_asesinatos.head(5).index,num_asesinatos.head(5),palette='coolwarm')
```

**7.Determinar las provincias con menor indice de homicidios y asesinatos.**

```python
num_asesinatos = datos['Numero_asesinatos'].groupby(datos.Provincia).mean().sort_values(ascending=False)
f,ax = plt.subplots(1,1,figsize=(15,10))
ax = sns.barplot(num_asesinatos.tail(5).index,num_asesinatos.tail(5),palette='coolwarm')
```

**8.Comprobar el tipo de muerte que mas se ha llevado a cabo durante este año.**

```python
p = sns.barplot(x='tipo_muert_matriz', y='Numero_asesinatos', data=datos)
```

**10.Determinar el tipo de muerte y comparar entre los estados civiles de las victimas**

```python
f,ax = plt.subplots(1,1,figsize=(15,10))
ax = sns.barplot(x='tipo_muert_matriz', y='Numero_asesinatos', hue='Estado Civil', data=datos)
```

**11.Determinar la edad con mayores victimas de homicidios y asesinatos.**
**12.Determinar la edad con menor victimas de homicidios y asesinatos.**

```python
fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(1, 1, 1)
ax.plot(datos.groupby(['Edad']).Numero_asesinatos.sum(),'ko--')
tick = ax.set_xticks(range(0,80,10))
ax.set_title('Total Homicidios y Asesinatos de 0 a 80 años')
ax.set_xlabel('Edad')
ax.set_ylabel('Numero_asesinatos')
```

**13.Analizar la motivacion para cometer los delitos de homicidios y asesinatos**

```python
serie_count_muertes=datos.Presunta_Motivacion.value_counts()
f,ax = plt.subplots(1,1,figsize=(15,10))
ax = sns.barplot(serie_count_muertes,serie_count_muertes.index,palette='coolwarm')
```

**14.Determinar los motivos para el mayor numero de Asesinatos y Homicidios**


```python
fig = plt.figure(figsize=(15,8))
serie_count_muertes=datos.Presunta_Motivacion.value_counts().head(5)
patches,texts=plt.pie(serie_count_muertes,startangle=90,shadow=True)
plt.pie(serie_count_muertes,shadow=True,labels=serie_count_muertes.index,autopct='%1.1f%%')
plt.legend(patches,serie_count_muertes.index,loc="best")
plt.axis('equal')
plt.title("Mayores motivos para Asesinatos y Homicidios")
plt.tight_layout()
plt.show()
```

**15.Determinar los motivos para el menor numero de Asesinatos y Homicidios**


```python
fig = plt.figure(figsize=(15,8))
serie_count_muertes=datos.Presunta_Motivacion.value_counts().tail(5)
patches,texts=plt.pie(serie_count_muertes,startangle=90,shadow=True)
plt.pie(serie_count_muertes,shadow=True,labels=serie_count_muertes.index,autopct='%1.1f%%')
plt.legend(patches,serie_count_muertes.index,loc="best")
plt.axis('equal')
plt.title("Mayores motivos para Asesinatos y Homicidios")
plt.tight_layout()
plt.show()
```

**16.Determinar el tipo de arma que se utiliza con mayor frecuencia.**

```python
fig = plt.figure(figsize=(15,8))
serie_count_muertes_sin_otros=datos.Tipo_Arma.value_counts().head(5)
patches,texts=plt.pie(serie_count_muertes_sin_otros,startangle=90,shadow=True)
plt.pie(serie_count_muertes_sin_otros,shadow=True,labels=serie_count_muertes_sin_otros.index,autopct='%1.1f%%')
plt.legend(patches,serie_count_muertes_sin_otros.index,loc="best")
plt.axis('equal')
plt.title("Top 5 de Armas Usadad")
plt.tight_layout()
plt.show()
```

**17.Determinar las horas con mayor numero de Asesinatos y Homicidios**

```python
serie_count_muertes_sin_otros=datos.Hora_Infraccion.value_counts().head(5)
plt.pie(serie_count_muertes_sin_otros,shadow=True,labels=serie_count_muertes_sin_otros.index,autopct='%1.1f%%')
plt.legend(patches,serie_count_muertes_sin_otros.index,loc="best")
plt.axis('equal')
plt.title("Hora con mayor numero de asesinatos y homicidios")
plt.tight_layout()
plt.show()
```

**18.Determinar las horas con menor numero de Asesinatos y Homicidios**

```python
serie_count_muertes_sin_otros=datos.Hora_Infraccion.value_counts().tail(5)
plt.pie(serie_count_muertes_sin_otros,shadow=True,labels=serie_count_muertes_sin_otros.index,autopct='%1.1f%%')
plt.legend(patches,serie_count_muertes_sin_otros.index,loc="best")
plt.axis('equal')
plt.title("Hora con menos numero de asesinatos y homicidios")
plt.tight_layout()
plt.show()
```
