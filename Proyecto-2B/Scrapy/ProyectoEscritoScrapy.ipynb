**Analizar los precios de los precios de Samsung**

```python
df_Samsung["FECHA LANZAMIENTO"]=pd.to_datetime(df_Samsung["FECHA LANZAMIENTO"])
phone_date = df_Samsung.sort_values(by = ['FECHA LANZAMIENTO'],ascending=False)
columns=filtered_columns = ['MODELO','FECHA LANZAMIENTO','PRECIOS']

phone_filtered = phone_date.reindex(columns=filtered_columns)
phone_filtered.head(30)
```

**Relacion de la Fecha de Lanzamiento y el Precio de Samsung**

```python
plt.figure(1,figsize=(20,5))
plt.plot(phone_date['FECHA LANZAMIENTO'], phone_date['PRECIOS'])
plt.suptitle('Precio Inicial vs Fecha de Lanzamiento') 
plt.show()
```

**Analizar los precios de los precios del Iphone**

```python
df_Iphone["FECHA LANZAMIENTO"]=pd.to_datetime(df_Iphone["FECHA LANZAMIENTO"])
phone_date1 = df_Iphone.sort_values(by = ['FECHA LANZAMIENTO'],ascending=False)
columns=filtered_columns = ['MODELO','FECHA LANZAMIENTO','PRECIOS']

phone_filtered1 = phone_date1.reindex(columns=filtered_columns)
phone_filtered1.head(30)
```

**Relacion de la Fecha de Lanzamiento y el Precio del Iphone**

```python
plt.figure(1,figsize=(20,5))
plt.plot(phone_date1['FECHA LANZAMIENTO'], phone_date1['PRECIOS'])
plt.suptitle('Precio Inicial vs Fecha de Lanzamiento') 
plt.show()
```

**Relacionar la RAM en cadad modelo de los telefonos Samsung.**

```python
modelo_s = df_Samsung['MODELO']
ram_s = df_Samsung['RAM (GB)']
plt.figure(1,figsize=(20,5))
plt.xticks(rotation = 'vertical')
plt.bar(modelo_s,ram_s,color = 'blue')
```
